# **To create an AI-powered assistant that educates users about diseases in their local language using voice, visuals, and text.**

```python

from googletrans import Translator
from gtts import gTTS
import wikipedia
from IPython.display import Audio, Image, display
from youtube_transcript_api import YouTubeTranscriptApi

# inputs disease name
disease = input("Enter a disease name: ")

# Translate disease name to Tamil (or change 'ta' to another language code)
translator = Translator()
translated = translator.translate(disease, dest='ta')
print("Tamil Translation:", translated.text)

# Get description from Wikipedia
wikipedia.set_lang("en")
try:
    description = wikipedia.summary(disease, sentences=2)
    print("\nüìÑ Wikipedia Description:\n", description)
except:
    description = "No Wikipedia description found."
    print("\n‚ùå Could not fetch description from Wikipedia.")

#Convert description to audio
tts = gTTS(description)
tts.save("disease.mp3")
display(Audio("disease.mp3"))

# Show image from Wikipedia (if available)
try:
    page = wikipedia.page(disease)
    image_url = None
    for img in page.images:
        if img.lower().endswith((".jpg", ".jpeg", ".png")) and "logo" not in img.lower():
            image_url = img
            break
    if image_url:
        print("\nüñºÔ∏è Showing related image from Wikipedia:")
        display(Image(url=image_url))
    else:
        print("No suitable image found.")
except:
    print("‚ùå Could not retrieve image from Wikipedia.")

```
